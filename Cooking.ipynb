```python
#Importing Python packages
import pandas as pd
import numpy as np
```

```python
#Getting the trainging dataset info
train_data = pd.read_json('train.json')
cuisine_set = set()
ingredients_set = set()
for item in train_data['cuisine']:
    cuisine_set.add(item)
cuisine_set = list(cuisine_set)
for item in train_data['ingredients']:
    for i in item:
        ingredients_set.add(i)
ingredients_set = list(ingredients_set)
ingredients_set.sort()
```

```python
#Analyzing the training dataset
print(train_data.info())
print(train_data['cuisine'].value_counts())
```

```python
#Processing the train data
train_lst = []
for ing_list in train_data['ingredients']:
    lst = []
    for item in ingredients_set:
        if item in ing_list:
            lst.append(1)
        else:
            lst.append(0)
    train_lst.append(lst)

df = pd.DataFrame(train_lst)
df.columns = ingredients_set
```

```python
#Generating Training Data
X = np.array(df)

# Import label encoder 
from sklearn import preprocessing 

# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# Encode labels in column 'cuisine'. 
train_data['cuisine']= label_encoder.fit_transform(train_data['cuisine']) 

Y = np.array(train_data['cuisine'])
```

```python
#Training the Classifier
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X, Y)
```

```python
#Getting the test dataset info
test_data = pd.read_json('test.json')
```

```python
#Processing the test data
test_lst = []
for ing_list in test_data['ingredients']:
    lst = []
    for item in ingredients_set:
        if item in ing_list:
            lst.append(1)
        else:
            lst.append(0)
    test_lst.append(lst)

df = pd.DataFrame(test_lst)
df.columns = ingredients_set
test = np.array(df)
```

```python
#Predicting the answers
pred = clf.predict(test)
print(pred)
ans = label_encoder.inverse_transform(pred)
print(ans)
```

```python
for i in ans:
    print(i)
```

```python
df1 = test_data['id']
df2 = pd.DataFrame(ans)
df2.columns = ['cuisine']
df = pd.concat([df1,df2], axis=1)
print(df.head())
df.to_csv('ans.csv')
```
